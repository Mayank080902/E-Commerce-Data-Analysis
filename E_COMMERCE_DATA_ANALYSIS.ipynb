Standardize Column Names

```python
data.columns = [
    'user_id', 'product_id', 'category', 'price_rs',
    'discount_pct', 'sales_revenue', 'payment_method', 'purchase_date'
]
```

Duplicates and Missing Values

```python
initial_rows = len(data)
data.drop_duplicates(inplace=True)
duplicates_removed = initial_rows - len(data)
```

```python
print(f"{duplicates_removed} duplicate rows removed.")
```

```python
# Convert date string to datetime object (format is DD-MM-YYYY)


data['purchase_date'] = pd.to_datetime(data['purchase_date'], format='%d-%m-%Y', errors='coerce')
data.dropna(subset=['purchase_date'], inplace=True) # Drop rows if date conversion failed
```

```python
# Create time-based features
data['purchase_month'] = data['purchase_date'].dt.to_period('M').astype(str)
data['day_of_week'] = data['purchase_date'].dt.day_name()
data['purchase_year'] = data['purchase_date'].dt.year
```

```python
data
```

```python
# Data Integrity Check (Price Consistency)

# Verify if 'sales_revenue' is consistent with 'price_rs' * (1 - 'discount_pct')

data['calculated_price'] = data['price_rs'] * (1 - data['discount_pct'] / 100)
price_difference = (data['sales_revenue'] - data['calculated_price']).abs()
inconsistent_prices = (price_difference > 1).sum()
print(f"Number of price inconsistencies (diff > 1 Rs): {inconsistent_prices}")
```

```python
data_clean = data.drop(columns=['price_rs', 'discount_pct', 'calculated_price'])
print("\nCleaned Data Summary:")
print(data_clean.info())
```

RFM analysis - Customer segmentation

```python
# Define a reference date (one day after the last transaction date)
snapshot_date = data['purchase_date'].max() + pd.Timedelta(days=1)

# Calculate RFM metrics
rfm_data = data.groupby('user_id').agg(
    recency=('purchase_date', lambda x: (snapshot_date - x.max()).days),
    frequency=('product_id', 'count'),
    monetary=('sales_revenue', 'sum')
).reset_index()


# R: Lower recency is better (higher score).
rfm_data['R_Score'] = pd.qcut(rfm_data['recency'], 4, labels=[4, 3, 2, 1], duplicates='drop')


# M: Higher monetary is better (higher score).
rfm_data['M_Score'] = pd.qcut(rfm_data['monetary'], 4, labels=[1, 2, 3, 4], duplicates='drop')


# Combine scores
rfm_data['RFM_Score'] = rfm_data['R_Score'].astype(str) + '1' + rfm_data['M_Score'].astype(str)
```

```python
def rfm_segment(score):
    if score.startswith('4') and score.endswith('4'): # High R, High M
        return 'Champion Customers'
    elif score.startswith('1') and score.endswith('1'): # Low R, Low M
        return 'Lost Customers'
    else:
        return 'Other Segment'

rfm_data['Segment'] = rfm_data['RFM_Score'].apply(rfm_segment)

segment_counts = rfm_data['Segment'].value_counts().sort_values(ascending=False)
```

Top Performing Category & Discount Impact

```python
# Category Revenue and Discount Analysis

# 1. Calculate Revenue by Category
category_revenue = data.groupby('category')['sales_revenue'].sum().sort_values(ascending=False).reset_index()

top_5_categories = category_revenue.head(5)['category'].tolist()

category_revenue_top5 = category_revenue.head(5)

# 2. Calculate Average Discount for Top 5 Categories
top5 = data[data['category'].isin(top_5_categories)]

avg_discount = top5.groupby('category')['discount_pct'].mean().reset_index()

avg_discount.rename(columns={'discount_pct': 'avg_discount_pct'}, inplace=True)

insight_2_data = pd.merge(category_revenue_top5, avg_discount, on='category')

print("\nInsight 2 - Top Categories and Avg Discount:")
print(insight_2_data)
```

```python
# Visualization
plt.figure(figsize=(8, 4))
sns.barplot(x='category', y='sales_revenue', data=insight_2_data, palette='viridis')
plt.title('Top 5 Categories by Sales Revenue')
plt.xlabel('Category')
plt.ylabel('Total Sales Revenue (Rs.)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='category', y='avg_discount_pct', data=insight_2_data, palette='viridis')
plt.title('Average Discount Percentage for Top 5 Categories')
plt.xlabel('Category')
plt.ylabel('Average Discount (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Seasonal Sales Trend

Monthly Sales Trend

```python
monthly_trends = data.groupby('purchase_month').agg(
    total_revenue=('sales_revenue', 'sum'),
    total_orders=('user_id', 'count')
).reset_index()

# Sort by time period
monthly_trends.sort_values(by='purchase_month', inplace=True)
print("\nInsight 3 - Monthly Sales Trends (Top 3 Months):")
print(monthly_trends.sort_values(by='total_revenue', ascending=False).head(3))
```

```python
# Visualization

plt.figure(figsize=(10, 5))
sns.lineplot(x='purchase_month', y='total_revenue', data=monthly_trends, marker='o')
plt.title('Monthly Sales Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales Revenue (Rs.)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x='purchase_month', y='total_orders', data=monthly_trends, marker='o', color='orange')
plt.title('Monthly Order Volume Trend')
plt.xlabel('Month')
plt.ylabel('Total Orders')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Average Sales Revenue by Payment Method

AOV - Average Order Value by Payment Method

```python
#  AOV by Payment Method

# Calculate Average Sales Revenue (AOV Proxy) by Payment Method

payment_aov = data.groupby('payment_method')['sales_revenue'].mean().sort_values(ascending=False).reset_index()
payment_aov.rename(columns={'sales_revenue': 'avg_sales_revenue'}, inplace=True)
print("\nInsight 4 - AOV by Payment Method:")
print(payment_aov)
```

```python
# Visualization

plt.figure(figsize=(8, 4))
sns.barplot(x='payment_method', y='avg_sales_revenue', data=payment_aov, palette='viridis')
plt.title('Average Sales Revenue by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Average Sales Revenue (Rs.)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Customer Loyalty Profile (Daily Activity)

```python
# --- Script Excerpt: Daily Sales Activity ---
# Calculate Total Order Volume by Day of Week
day_counts = data.groupby('day_of_week')['user_id'].count().reset_index()
day_counts.rename(columns={'user_id': 'total_orders'}, inplace=True)

# Define order for days of week for proper sorting in report
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_counts['day_of_week'] = pd.Categorical(day_counts['day_of_week'], categories=day_order, ordered=True)
day_counts.sort_values('day_of_week', inplace=True)
print("\nInsight 5 - Order Volume by Day of Week:")
print(day_counts)
```

```python
# Visualization

plt.figure(figsize=(10, 6))
sns.barplot(x='day_of_week', y='total_orders', data=day_counts, palette='viridis')
plt.title('Order Volume by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Total Orders')
plt.tight_layout()
plt.show()
```
